```python
import numpy as np 
from scipy.stats import sem, t
from numpy import mean
confidence = 0.05
data = BNP
n = len(data)
m = mean(data)
std_err = sem(data)
h = std_err * t.ppf((1+ confidence) / 2, n-1)

start = m-h
end = m+h
print ("95% Confidence Interval of BNP is from ", start, " to ", end)


```

# convert Gender into integers

```python
import numpy as np 
import pandas as pd
import scipy as py

#convert string values to integers 0 and 1
sw['Gender'] = sw['Gender'].map({'M': 1, 'F':0})

sw
```

```python
import matplotlib.pyplot as plt #import matplot library and label as plt

#data stream is sw here

labels = 'Male', 'Female'
sizes = [sw.loc[sw['Gender'] == 1].shape[0], sw.loc[sw['Gender'] == 0].shape[0]]
explode = (0, 0.1)
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)

plt.show()
```

```python
labels = 'Glasses','Contact Lens','Both', 'Lasik' #label on pie chart
sizes = [sw.loc[sw['Changing'] == 1].shape[0], sw.loc[sw['Changing'] == 2].shape[0], sw.loc[sw['Changing'] == 3].shape[0], sw.loc[sw['Changing'] == 4].shape[0]]
#tally up each number. total number of 1s, 2s, 3s, and 4s

explode = (0, 0.1, 0, 0) # only explode contact 


plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode = explode)

plt.show()

```

```python
# Easy Pie Graph. If you know numbers already
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Glasses', 'Contact Lens', 'Both', 'Lasik'
sizes = [12, 23, 63, 2]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Contact Lens')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
